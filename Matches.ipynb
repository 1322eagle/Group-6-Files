```python
#removing null values 


ds['umpire3']=ds['umpire3'].fillna('digital')
ds['city']=ds['city'].fillna('data not avl.')
ds['player_of_match']=ds['player_of_match'].fillna('No data')
ds['winner']=ds['winner'].fillna('No data')
ds['umpire1']=ds['umpire1'].fillna('RJ Tucker')
ds['umpire2']=ds['umpire2'].fillna('HDPK Dharmasena')

ds['result'] = ds['result'].replace('no result', 'tie')
```

```python
rg=ds.groupby("team2")["result"].value_counts()
rg
```

```python
df=ds.groupby("team2")["result"].value_counts()
df_unstacked = df.unstack(level=-1)
fig, ax = plt.subplots(figsize=(8, 6))

colors=["#9080ff","#ffb400"]
df_unstacked.plot(kind='bar', stacked=True, ax=ax,color=colors)
ax.set_xlabel('Teams ')
ax.set_ylabel('Number of Matches')
ax.set_title('Number of Normal Wins and Ties for Teams in group 2')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
